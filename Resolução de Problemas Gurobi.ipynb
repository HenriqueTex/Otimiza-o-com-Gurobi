```python
import gurobipy as gp
```

```python
#Criando modelo
m= gp.Model("Problema do Sapateiro")

#Variaveis de decisão
x1=m.addVar()
x2=m.addVar()

#Função Objetivo
m.setObjective(5*x1 + 2*x2, sense=gp.GRB.MAXIMIZE)

#Restrições 
c1=m.addConstr(2*x1 + x2<=6)
c2=m.addConstr(10*x1 + 12*x2<=60)

#Resolvendo o modelo
m.optimize()

#Mix de produção
print("Sapatos: ",x1.X)
print("Cintos: ",x2.X)

#Material utilizado utilizado
print ("A quantidade de couro utlizada foi ",6-c1.Slack)

#Tempo utilizado
print ("A quantidade de tempo utilizada foi", 60-c2.Slack, " minutos")
```
